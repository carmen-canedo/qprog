```python
# Task 1
def inner_product(circuit, a):
    n = len(a)
    
    for i in range(len(a)):
        if a[::-1][i] == "1":
            circuit.cx(i, n)

    circuit.barrier()
```

```python
# Task 1 check
a = "01101"
n = len(a)
circuit = QuantumCircuit(n+1, 0)
inner_product(circuit, a)
print(circuit)
```

```python
# Task 2
def hadamards(circuit):
    for i in range(circuit.num_qubits):
        circuit.h(i)
    circuit.barrier()
```

```python
# Task 2 check
circuit = QuantumCircuit(5, 0)
hadamards(circuit)
print(circuit)
```
